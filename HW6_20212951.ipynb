```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = range(1, 50)
Y = [value * 3 for value in X]
plt.xlim(0,50)
plt.ylim(0,160)
plt.plot(X, Y, 'b', linewidth=2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Draw a line')
plt.show()

print("Values of X:")
print(*range(1,50))
print("Values of Y (thrice of X):")
print(Y)
```

```python
x1 = [10,20,30]
y1 = [20,40,10]

x2 = [10,20,30]
y2 = [40,10,30]
plt.xlim(10,30)
plt.ylim(10,40)
plt.plot(x1, y1, 'b',label='line 1')
plt.plot(x2, y2, 'g', label='line 2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title("Two or more lines on same plot with suitable legends")
plt.show()
```

```python
x1 = [10,20,30]
y1 = [20,40,10]

x2 = [10,20,30]
y2 = [40,10,30]
plt.xlim(10,30)
plt.ylim(10,40)
plt.plot(x1, y1, 'b', linewidth=3, label='line 1-width-3')
plt.plot(x2, y2, 'r', linewidth=5, label='line 2-width-5')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title("Two or more lines with different widths and colors with suitable legends")
plt.show()
```

```python
x1 = [10,20,30]
y1 = [20,40,10]

x2 = [10,20,30]
y2 = [40,10,30]
plt.xlim(10,30)
plt.ylim(10,40)
plt.plot(x1, y1, 'b:', linewidth=1, label='line 1-dotted')
plt.plot(x2, y2, 'r:', linewidth=5, label='line 2-dashes')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title("Plot with two or more lines with different style")
plt.show()
```

```python
x = [1,4,5,6,7]
y = [2,6,3,6,3]

plt.plot(x, y, 'r-.', linewidth=1)
plt.plot(x, y, 'bo', linewidth=3)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(1,8)
plt.ylim(1,8)
plt.title("Display marker")
plt.show()
```

```python
t = np.arange(0., 5., 0.2)

plt.xlim(0,5)
plt.ylim(0,120)
plt.plot(t, t, 'g--')

plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'r^')
plt.show()
```

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]

plt.bar(x,popularity, align='edge',color='b')
plt.xlim(0,6)
plt.ylim(0,25)
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago")
plt.xticks(x_pos, x)

plt.minorticks_on()
plt.grid(color='r')
plt.grid(which='minor',color='gray', linestyle='--',linewidth=0.5)

plt.show()
```

```python
n_groups = 5
men_means = (22, 30, 33, 30, 26)
women_means = (25, 32, 30, 35, 29)


fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

plt.bar(index, men_means, color='g',alpha=opacity, linewidth=0.8, edgecolor='k', width=bar_width, align='edge', label='Men')
plt.bar(index+bar_width, women_means, color='r',alpha=opacity,linewidth=0.8,edgecolor='k',width=bar_width,align='edge',label='Women')

plt.xlim(0,5)
plt.ylim(0,35)
plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()
plt.tight_layout()
plt.show()
```

```python
N = 5
menMeans = (54.74, 42.35, 67.37, 58.24, 30.25)
menStd = (4, 3, 4, 1, 5)

ind = np.arange(N)    

width = 0.35   

plt.bar(ind, menMeans, color='r',linewidth=0.8, edgecolor='k',width=width, align='edge')
plt.errorbar(ind+width/2, menMeans, yerr=menStd, linestyle='', elinewidth=0.5, color='b', capsize=2)

plt.ylabel('Scores')
plt.xlabel('Velocity')
plt.title('Scores by Velocity')
plt.xlim(0.0,4.5)
plt.ylim(0,80)

plt.minorticks_on()
plt.grid(color='g', linewidth=0.5)
plt.grid(which='minor',color='gray', linestyle='--',linewidth=0.5)
plt.show()
```
